```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img3 = cv2.imread('C:/Users/halil/Desktop/car16.jpg')
img2 = cv2.imread('C:/Users/halil/Desktop/cropped/1.jpg')

img = cv2.imread('C:/Users/halil/Desktop/deneme.png')
```

```python
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
gray = cv2.bilateralFilter(gray, 11, 17, 17)  # 11  //TODO 11 FRO OFFLINE MAY NEED TO TUNE TO 5 FOR ONLINE

gray = cv2.medianBlur(gray, 5)
edged = cv2.Canny(gray, 30, 400)
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
plt.show()

```

```python
countours, _= cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
```

```python
imageCopy = img2.copy()
a = cv2.drawContours(imageCopy, countours, -1, (0, 255, 0), 1)
plt.imshow(cv2.cvtColor(a, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cnts = sorted(countours, key=cv2.contourArea, reverse=True)
screenCntList = []
scrWidths = []
for cnt in cnts:
    peri = cv2.arcLength(cnt, True)  # cnts[1] always rectangle O.o
    approx = cv2.approxPolyDP(cnt, 0.02 * peri, True)
    screenCnt = approx
    # print(len(approx))

    if (len(screenCnt) == 4):

        (X, Y, W, H) = cv2.boundingRect(cnt)
        # print('X Y W H', (X, Y, W, H))
        screenCntList.append(screenCnt)
        scrWidths.append(W)

    else:
        print("4 points not found")

print('Screens found :', len(screenCntList))
print('Screen Dimentions', scrWidths)

#creenCntList, scrWidths = findLargestCountours(screenCntList, scrWidths)
```
